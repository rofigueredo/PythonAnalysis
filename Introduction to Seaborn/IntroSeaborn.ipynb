```python
# Import pandas
import pandas as pd

# Create a DataFrame from csv file
df = pd.read_csv('young-people-survey-responses.csv')

# Print the head of df
df.head()
```

```python
# Import Matplotlib, pandas, and Seaborn using the standard names.
import matplotlib.pyplot as plt
import seaborn as sns

sns.countplot(x='Spiders', data=df)

plt.show()
```

```python
student_data = pd.read_csv('student-alcohol-consumption.csv')

sns.scatterplot(x="absences", y="G3", 
                data=student_data, 
                hue="location",
                hue_order=["Rural", "Urban"])

# Show plot
plt.show()
```

```python
# Create a dictionary mapping subgroup values to colors
palette_colors = {"Rural": "green", "Urban": "blue"}

# Create a count plot of school with location subgroups
sns.countplot(x="school", data=student_data,
              hue="location",
              palette=palette_colors)

# Display plot
plt.show()
```

```python
# Change this scatter plot to arrange the plots in rows instead of columns
sns.relplot(x="absences", y="G3", 
            data=student_data,
            kind="scatter",
            #col="study_time" 
            col="study_time")


# Show plot
plt.show()
```

```python
# Adjust further to add subplots based on family support
sns.relplot(x="G1", y="G3", 
            data=student_data,
            kind="scatter", 
            col="schoolsup",
            col_order=["yes", "no"],
            row='famsup',
            row_order=["yes", "no"])

# Show plot
plt.show()
```

```python
# Adjust to add subplots based on school support
sns.relplot(x="G1", y="G3", 
            data=student_data,
            kind="scatter", 
            col="schoolsup",
            col_order=["yes", "no"])

# Show plot
plt.show()
```

```python
# Create scatter plot of horsepower vs. mpg
mpg = sns.load_dataset('mpg')

sns.relplot(x="horsepower", y="mpg", 
            data=mpg, kind="scatter", 
            size="cylinders",
            hue='cylinders')

# Show plot
plt.show()
```

```python
# Create a scatter plot of acceleration vs. mpg
sns.relplot(x="acceleration", y="mpg", 
            data=mpg, kind="scatter", 
            style="origin", hue="origin")

# Show plot
plt.show()
```

```python
# Create line plot
sns.relplot(x="model_year", y="mpg",
            data=mpg, kind="line")

# Show plot
plt.show()
```

```python
# Add markers and make each line have the same style
sns.relplot(x="model_year", y="horsepower", 
            data=mpg, kind="line", 
            ci=None, style="origin", 
            hue="origin", markers=True,
            dashes=False)

# Show plot
plt.show()
```
